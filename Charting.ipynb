```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
fileList=glob.glob("stock_comment_count/*.csv")

for filename in fileList:
    stock_data_df=pd.read_csv(filename)
    ticker = filename.replace("stock_comment_count\\", "").replace(".csv", "")
    print(ticker)
    stock_data_df.head()


    # Create some mock data
    t = stock_data_df["Date"]
    stock_data = stock_data_df['Close']
    reddit_data = stock_data_df['comment_volume']

    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_xlabel('Date')
    ax1.set_ylabel(ticker + " Closing Price", color=color)
    ax1.plot(t, stock_data, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('Comment Volume', color=color)  # we already handled the x-label with ax1
    ax2.plot(t, reddit_data, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
```
